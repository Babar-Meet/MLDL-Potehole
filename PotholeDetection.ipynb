```python
from ultralytics import YOLO

MODEL_PATH = "/content/drive/MyDrive/pothole_yolo/train1/weights/last.pt"

import os

if os.path.exists(MODEL_PATH):
    print("Resuming previous training...")
    model = YOLO(MODEL_PATH)
    model.train(resume=True)
else:
    print("Starting fresh training...")
    model = YOLO("yolov8n.pt")
    model.train(
        data="/content/data.yaml",
        epochs=100,
        imgsz=640,
        batch=16,
        project="/content/drive/MyDrive/pothole_yolo",
        name="train1",
        exist_ok=True
    )
```

```python
BEST_MODEL = "/content/drive/MyDrive/pothole_yolo/train1/weights/best.pt"
```

```python
model = YOLO(BEST_MODEL)

metrics=model.val(data='/content/data.yaml')
print(metrics)
```

```python
model.predict(source="/content/test/images", conf=0.25, save=True)
```

```python
model.val(data="/content/data.yaml", plots=True)
```

```python
print("Precision:", metrics.results_dict['metrics/precision(B)'])
print("Recall:", metrics.results_dict['metrics/recall(B)'])
print("mAP50:", metrics.results_dict['metrics/mAP50(B)'])
print("mAP50-95:", metrics.results_dict['metrics/mAP50-95(B)'])
```
